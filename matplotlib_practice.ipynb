```python
#import dependencies for MATLAB
import matplotlib.pyplot as plt
```

```python
#set x-axis to list of strings for ea month
x_axis=['Jan','Feb','Mar','April','May','June','July','Aug','Sept','Oct','Nov','Dec']
y_axis=[10.02,23.24,39.20,35.42,32.34,27.04,43.82,10.56,11.85,27.90,20.71,20.09]
```

```python
#Create the plot
plt.plot(x_axis,y_axis)
```

```python
#create plot with ax.plt() object-oriented inteface
fig,ax=plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
#create plot w/ ax.plt()
fig=plt.figure()
ax=fig.add_subplot()
```

```python
#create plot w/ ax.plt() add reference
ax=plt.axes()
ax.plot(x_axis,y_axis)
```

```python
#create plot and add label for legend
plt.plot(x_axis,y_axis,marker='*',color='blue',linewidth=2,label='Boston')
#create labels for x and y axes
plt.xlabel('Date')
plt.ylabel('Fare($)')
#set y limit 0 - 45
plt.ylim(0,45)
#creat title
plt.title('PyBer Fare by Month')
#add legend
plt.legend()
```

```python
#plot chart and annotations using object-oriented interface
fig,ax=plt.subplots()
ax.plot(x_axis,y_axis,marker='D',color='g',linewidth=2,label='Boston')
ax.set_title('PyBer Fares by Month')
ax.set_xlabel('Month')
ax.set_ylabel('Fare($)')
ax.legend()
ax.set_ylim(0,55)
ax.grid()
```

```python
#create vertical bar chart plot
plt.bar(x_axis,y_axis,color='g',label='Boston')
#create labels for the axes
plt.xlabel("Date")
plt.ylabel('Fare($)')
#create title
plt.title("PyBer Fare by Month")
#add legend
plt.legend()
```

```python
#plot horizontal bar graph
plt.barh(x_axis,y_axis)
```

```python
#invert the y axes to ascending order
plt.barh(x_axis,y_axis,color='m',label='Boston')
plt.legend()
plt.title('PyBer Fare by Month')
plt.xlabel('Fare($)')
plt.ylabel('Month')
plt.gca().invert_yaxis()
```

```python
#create plot w/ ax.plt() for vertical bar chart
fig,ax=plt.subplots()
ax.bar(x_axis,y_axis)
```

```python
#create plot w/ ax.plt() for horizontal bar
fig,ax=plt.subplots()
ax.barh(x_axis,y_axis,color='c',label='Chicago')
ax.invert_yaxis()
ax.set_title("PyBer Fares by Month")
ax.set_xlabel('Fares($)')
ax.set_ylabel('Months')
ax.grid()
```

```python
#plot scatterplot
plt.plot(x_axis,y_axis,'o')
```

```python
#create scatter plot using plt.scatter()
plt.scatter(y_axis,x_axis,color='r',label='Chicago')
plt.xlim(0,50)
plt.title('PyBer Fares by Month')
plt.xlabel('Fares($)')
plt.ylabel('Month')
plt.gca().invert_yaxis()
plt.legend()
```

```python
#create a bubble chart
plt.scatter(x_axis,y_axis,s=y_axis)
```

```python
#make bubbles larger by iterating through y_axis data
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
#use new y_axis for s parameter
plt.scatter(x_axis,y_axis,s=y_axis_larger)
```

```python
#make bubbles larger using Python's list comprehension
plt.scatter(x_axis,y_axis,s=[i*3 for i in y_axis])
```

```python
#create scatter plot using object-oriented interface
fig,ax=plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
#bubble chart using object-oriented interface

#make markers 5x ea data point
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*5)

#format rest of chart    
fix,ax=plt.subplots()
ax.scatter(y_axis,x_axis,s=y_axis_larger,color='xkcd:sky blue',alpha=.8,edgecolors='black',label='Boston')
ax.legend()
ax.set_title('PyBer Fare by Month')
ax.invert_yaxis()
ax.set_xlabel('Fare($)')
ax.set_ylabel('Month')
ax.set_xlim(0,55)
```

```python
#pie chart
categories=['pants','shirts','outerwear','pajamas','underwear','shoes']
revenue=[25,15,5,20,10,25]
```

```python
#insert variable for size of each slice
plt.pie(revenue)
```

```python
#explode piece of pie
explode=[.1,0,0,0,0,0]
#aadd % up to 1 decimal point into pie chart
plt.pie(revenue,explode=explode,labels=categories,autopct='%1.1f%%')
plt.show()
```

```python
#Assign 12 colors, one for ea month and change fig size to make % fit better
colors=['slateblue','magenta','lightblue','green','yellowgreen','greenyellow','yellow','orange','gold','indianred','tomato','mistyrose']
explode_values=(0,0,0,0,0,0,.2,0,0,0,0,0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,explode=explode_values,colors=colors,labels=x_axis,autopct='%1.1f%%')

plt.show()
```
