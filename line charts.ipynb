```python
import matplotlib.pyplot as plt
```

```python
savings=[0,100,250,300,300,500,300,350,500,600,700,1000]
months=['Jan','Feb', 'Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
plt.plot(months, savings)
plt.xlabel('Month')
plt.ylabel('Savings ($)')
plt.title('Savings Account Results')
plt.show()
```
